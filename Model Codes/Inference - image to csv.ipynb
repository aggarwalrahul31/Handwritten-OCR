```python
#prediction test

from scipy.stats import mode
import pandas as pd
from PIL import Image,ImageFilter
import cv2
import numpy as np
import sys
import os
import csv
import pandas as pd
import PIL.ImageOps

columnNames=[]
for i in range (0,6084):
    columnNames.append(i)
data=pd.DataFrame(columns = columnNames)

myFileList=[]
for i in range(86,-1,-1):
    myFileList.append(r'C:\Users\Rahul\Desktop\Capstone\form\Inner\2\1'+'\\'+str(i)+'.jpg')

num_images=1

count1=0
for file in myFileList:
    print(file)
    image = Image.open(file)
    image = image.crop((4, 3, image.size[0]-4, image.size[0]-5))     # im.crop((left, top, right, bottom)) 
    img_resized = image.resize((78,78))
    img_resized=PIL.ImageOps.invert(img_resized)
    # img_file = img_resized.filter(ImageFilter.MedianFilter())
    threshold = 80
    img_file = img_resized.point(lambda p: p > threshold and 255)

    # Save Greyscale values
    value = np.asarray(img_file.getdata(), dtype=np.int).reshape((img_file.size[1], img_file.size[0]))
    value = value.flatten()
    
    if (mode(value[0:78])[0]>200 or (mode(value[0:78])[0] < 80 and mode(value[0:78])[1][0] < 54)):
        value[0:78]=0
    if (mode(value[-78:])[0]>200 or (mode(value[0:78])[0] < 80 and mode(value[0:78])[1][0] < 54)):
        value[-78:]=0
    if (mode(value[-156:-78])[0]>200 or (mode(value[0:78])[0] < 80 and mode(value[0:78])[1][0] < 54)):
        value[-156:-78]=0
        
        
    #print(value)  
    data.loc[num_images] = value
    num_images += 1
    count1+=1
data.to_csv(r"C:\Users\Rahul\Desktop\Capstone\OCR\colab data\inference_2_1.csv", index=False)
```
